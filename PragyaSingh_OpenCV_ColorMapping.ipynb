# Color Mapping
### In OpenCV, we can apply different color maps to an image using the method applyColorMap() of the class Imgproc. 
And, I will use Matplotlib library to display the picture.

```python
import cv2
import matplotlib.pyplot as plt
```

```python
im_gray = cv2.imread("picture.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(im_gray)
```

```python
im_color = cv2.applyColorMap(im_gray, cv2.COLORMAP_JET)
plt.imshow(im_color)
```

```python
im_color1 = cv2.applyColorMap(im_gray, cv2.COLORMAP_AUTUMN)
plt.imshow(im_color1)
```

```python
im_color2 = cv2.applyColorMap(im_gray, cv2.COLORMAP_BONE)
plt.imshow(im_color2)
```

```python
im_color3 = cv2.applyColorMap(im_gray, cv2.COLORMAP_COOL)
plt.imshow(im_color3)
```

```python
im_color4 = cv2.applyColorMap(im_gray, cv2.COLORMAP_HOT)
plt.imshow(im_color4)
```

```python
im_color5 = cv2.applyColorMap(im_gray, cv2.COLORMAP_HSV)
plt.imshow(im_color5)
```

```python
im_color6 = cv2.applyColorMap(im_gray, cv2.COLORMAP_JET)
plt.imshow(im_color6)
```

```python
im_color7 = cv2.applyColorMap(im_gray, cv2.COLORMAP_OCEAN)
plt.imshow(im_color7)
```
